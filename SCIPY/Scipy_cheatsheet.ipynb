# Scipy:
SciPy is a scientific computation library that uses NumPy underneath. SciPy stands for Scientific Python. It provides more utility functions for optimization, stats and signal processing. 
#### Scipy submodules:
| Sub-Module                       | Purpose                                                        |
| ---------------------------------|----------------------------------------------------------------|
| `scipy.constants`                | Physical and mathematical constants                           |
| `scipy.cluster`                  | Clustering algorithms                                          |
| `scipy.fft`                      | Fast Fourier transform (FFT) routines                          |
| `scipy.integrate`                | Numerical integration and ODE solvers                          |
| `scipy.interpolate`              | Interpolation techniques (splines, polynomial)                 |
| `scipy.io`                       | Input and output for various file formats                      |
| `scipy.linalg`                   | Extended linear algebra capabilities                            |
| `scipy.ndimage`                  | n-dimensional image processing                                 |
| `scipy.optimize`                 | Optimization algorithms (minimization, curve fitting)          |
| `scipy.signal`                   | Signal processing tools                                        |
| `scipy.sparse`                   | Sparse matrices and related algorithms                         |
| `scipy.spatial`                  | Spatial data structures and algorithms                          |
| `scipy.special`                  | Special mathematical functions                                  |
| `scipy.stats`                    | Statistical functions and distributions                        |
| `scipy.weave`                    | Integration of C/C++ code with Python                           |

Each of these sub-modules specializes in different aspects of scientific computing, allowing users to perform a wide range of tasks from optimization and signal processing to image analysis and statistical modeling.

## scipy.cluster

```python
from scipy import cluster
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import numpy as np
```

```python
dir(cluster)
```

## 1. Hierarchical Clustering

```python
# Generate sample data
np.random.seed(42)
data = np.random.rand(10, 2)

# Hierarchical clustering
linkage_matrix = linkage(data, method='ward')

# Dendrogram
plt.figure(figsize=(8, 6))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

## 2. K-Means Clustering

```python
from scipy.cluster.vq import kmeans

# Generate sample data
np.random.seed(42)
data = np.random.rand(10, 2)

# K-Means clustering
k = 3
centroids, _ = kmeans(data, k)

print("Data Points:", data)
print("K-Means Centroids:", centroids)
```

## 3. Vector Quantization

```python
from scipy.cluster.vq import vq, kmeans
np.random.seed(42)
data = np.random.rand(10, 2)

# K-Means clustering
k = 3
centroids, _ = kmeans(data, k)

# Vector quantization
clusters, _ = vq(data, centroids)

print("Data Points:", data)
print("K-Means Centroids:", centroids)
print("Vector Quantization Clusters:", clusters)
```

## Fast Fourier Transform (FFT):
fft(x, n=None, axis=-1, norm="backward"):

Performs a one-dimensional FFT on an array x.

n specifies the desired output size (padded with zeros if smaller).

Axis defines the dimension along which to perform the FFT.

Norm controls the scaling factor applied to the output (default: "backward" for backwards compatibility).

```python
from scipy.fft import fft

x = np.arange(8)
print(x)
fft_result = fft(x)

print(fft_result)
```

## Inverse FFT

```python
from scipy.fft import ifft
```

```python
ifft_result = ifft(fft_result)

print(ifft_result)
```

## fftfreq(n, d=1.0):

Calculates the corresponding frequency values for a given FFT output size n and sampling interval d.

```python
from scipy.fft import fftfreq,fftshift
freqs = fftfreq(8)

print(freqs)  
```

```python
shifted_spectrum = fftshift(fft_result)

# Plot the shifted spectrum
import matplotlib.pyplot as plt

plt.plot(freqs, np.abs(shifted_spectrum))
plt.show()
```

## 3. Multi-dimensional FFT


```python
import numpy as np
from scipy.fft import fftn
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate 3D sample data
x = np.linspace(0, 4, 100)
y = np.linspace(0, 4, 100)
z = np.linspace(0, 4, 100)
X, Y, Z = np.meshgrid(x, y, z)
signal = np.sin(X) * np.cos(Y) * np.sin(Z)

# Perform 3D FFT
fft_result = np.array(fftn(signal))
print(fft_result.shape)

print(fft_result)
```

## Minimization:

minimize: Minimizes a scalar function of one or more variables using various optimization methods.

minimize_scalar: Minimizes a scalar function of one variable using a variety of algorithms.

```python
from scipy.optimize import minimize

# Define the objective function
def objective(x):
    return x**2 + 10*np.sin(x)

# Perform minimization
result = minimize(objective, x0=0)

print("Minimum value:", result.fun)
print("Optimal solution:", result.x)
```

## Curve Fitting:

curve_fit: Fits a function to data using non-linear least squares.

```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt

# Define the model function
def model_function(x, a, b):
    return a * x + b

# Generate sample data
x_data = np.array([1, 2, 3, 4, 5])
y_data = 2 * x_data + 1 + np.random.normal(scale=0.1, size=len(x_data))

# Perform curve fitting
params, covariance = curve_fit(model_function, x_data, y_data)

print("Optimal parameters:", params)

# Plot the original data and the fitted curve
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, model_function(x_data, *params), label='Fitted Curve', color='red')
plt.legend()
plt.show()
```

## Root Finding:

root: Finds a root of a vector function.

```python
from scipy.optimize import root

# Define the vector function
def vector_function(x):
    return [x[0]**2 + x[1]**2 - 1, x[0] - x[1]]

# Initial guess
initial_guess = [0.5, 0.5]

# Perform root finding
result = root(vector_function, initial_guess)

print("Root:", result.x)
```

## Global Optimization:

basinhopping: Finds the global minimum of a scalar function.


```python
from scipy.optimize import basinhopping

# Define the objective function
def objective(x):
    return x**2 + 10*np.sin(x)

# Perform global optimization
result = basinhopping(objective, x0=0)

print("Global minimum value:", result.fun)
print("Optimal solution:", result.x)
```

# scipy.signal
## 1. Convolution and Correlation:

```python
import numpy as np
from scipy.signal import convolve
import matplotlib.pyplot as plt

# Define two 1D arrays
signal = np.array([1, 2, 3, 4])
kernel = np.array([0.5, 0.5])

# Perform convolution
result = convolve(signal, kernel, mode='full')

# Plot the original signal and the convolved result
plt.stem(signal, label='Original Signal', basefmt='b-')
plt.stem(result, label='Convolved Result', basefmt='r-')
plt.legend()
plt.show()
```

## Filtering:

```python
from scipy.signal import lfilter
import numpy as np
import matplotlib.pyplot as plt

# Generate a noisy signal
t = np.linspace(0, 1, 100)
signal = np.sin(2 * np.pi * 5 * t) + 0.2 * np.random.randn(100)

# Design a low-pass filter
b = [0.01] * 10  # Numerator coefficients
a = [1.0]  # Denominator coefficients

# Apply the filter using lfilter
filtered_signal = lfilter(b, a, signal)

# Plot the original and filtered signals
plt.plot(t, signal, label='Original Signal')
plt.plot(t, filtered_signal, label='Filtered Signal', linewidth=2)
plt.legend()
plt.show()
```

## Spectral Analysis:

```python
from scipy.signal import periodogram
import numpy as np
import matplotlib.pyplot as plt

# Generate a sinusoidal signal
t = np.linspace(0, 1, 1000, endpoint=False)
signal = np.sin(2 * np.pi * 5 * t)

# Compute the power spectral density using periodogram
frequencies, psd = periodogram(signal)

# Plot the power spectral density
plt.semilogy(frequencies, psd)
plt.title('Power Spectral Density')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power/Frequency (dB/Hz)')
plt.show()
```

## Wavelet Transforms:

```python
from scipy.signal import cwt, morlet
import numpy as np
import matplotlib.pyplot as plt

# Generate a signal with a frequency that changes over time
t = np.linspace(0, 1, 1000, endpoint=False)
signal = np.sin(2 * np.pi * (10 + 5 * t) * t)

# Perform continuous wavelet transform (CWT)
widths = np.arange(1, 31)
cwt_result = cwt(signal, morlet, widths)

# Plot the original signal and its wavelet transform
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')

plt.subplot(2, 1, 2)
plt.imshow(np.abs(cwt_result), aspect='auto', extent=[0, 1, 1, 31], cmap='jet', interpolation='bilinear')
plt.title('Continuous Wavelet Transform')
plt.xlabel('Time')
plt.ylabel('Scale')
plt.show()
```

# scipy.stats
## 1. Descriptive Statistics:

```python
from scipy.stats import describe
import numpy as np

# Generate a random sample
data = np.random.normal(loc=5, scale=2, size=100)

# Calculate descriptive statistics
stats = describe(data)

print("Mean:", stats.mean)
print("Variance:", stats.variance)
print("Skewness:", stats.skewness)
print("Kurtosis:", stats.kurtosis)
```

## 2. Probability Distributions:

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Generate random samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Fit a normal distribution to the data
mu, std = norm.fit(data)

# Plot the histogram and the fitted normal distribution
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
plt.title("Fit results: mu = %.2f,  std = %.2f" % (mu, std))
plt.show()
```

## 3. Hypothesis Testing:

```python
from scipy.stats import ttest_ind
import numpy as np

# Generate two independent samples
group1 = np.random.normal(loc=5, scale=2, size=100)
group2 = np.random.normal(loc=8, scale=2, size=100)

# Perform independent samples t-test
t_stat, p_value = ttest_ind(group1, group2)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

## 4. Continuous Random Variables:

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generate random variates from a normal distribution
data = norm.rvs(loc=0, scale=1, size=1000)

# Plot the histogram of the random variates
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Random Variates from a Normal Distribution')
plt.show()
```

## scipy.interpolate

## 1. Linear Interpolation:

```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt

# Generate a set of data points
x_data = np.linspace(0, 10, 11)
y_data = np.sin(x_data)

# Create a linear interpolation function
linear_interp = interp1d(x_data, y_data, kind='linear')

# Generate new data points for interpolation
x_interp = np.linspace(0, 10, 101)
y_interp = linear_interp(x_interp)

# Plot the original data and linear interpolation
plt.scatter(x_data, y_data, label='Original Data')
plt.plot(x_interp, y_interp, label='Linear Interpolation', color='red')
plt.legend()
plt.show()
```

## Spline Interpolation:

```python
from scipy.interpolate import splrep, splev
import numpy as np
import matplotlib.pyplot as plt

# Generate a set of data points
x_data = np.linspace(0, 10, 11)
y_data = np.sin(x_data)

# Create a cubic spline representation
spline_representation = splrep(x_data, y_data, k=3)

# Evaluate the spline at new data points
x_interp = np.linspace(0, 10, 101)
y_interp = splev(x_interp, spline_representation)

# Plot the original data and spline interpolation
plt.scatter(x_data, y_data, label='Original Data')
plt.plot(x_interp, y_interp, label='Cubic Spline Interpolation', color='red')
plt.legend()
plt.show()
```
