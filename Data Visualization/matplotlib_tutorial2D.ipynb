```python
s_class=[9,10,11,12]
boys=[60,50,90,85]
girls=[45,45,100,90]

bar1=[i+0.4 for i in s_class]
bar2=[i+0.2 for i in s_class]


plt.bar(s_class,boys,0.4,label="Boys")
plt.bar(bar1,girls,0.4,label="Girls")

plt.xticks(bar2,s_class)

plt.title("Student vs class")
plt.xlabel("Class")
plt.ylabel("No. of Students")
plt.legend()
plt.show()
```

# Subplot

subplot() function you can draw multiple plots in one figure. 
## Syntax:
subplot(row,column,_plot)
## Example:
*subplot(1,3,1)* --> 1X3=3 elements

This means there are 1 row 3 column 1st plot 

*subplot(1,3,2)*

This means there are 1 row 3 column 2nd plot 

*subplot(1,3,3)*

This means there are 1 row 3 column 3rd plot 

```python
#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 3, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 3, 2)
plt.plot(x,y)

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([15, 25, 35, 45])

plt.subplot(1, 3, 3)
plt.plot(x,y)

plt.show()
```

```python
#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(3, 1, 2)
plt.plot(x,y)

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([15, 25, 35, 40])

plt.subplot(3, 1, 3)
plt.plot(x,y)

plt.show()
```
