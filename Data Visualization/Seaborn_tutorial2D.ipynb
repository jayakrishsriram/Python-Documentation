```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips") 
```

```python
df.head()
```

```python
#sns.set(style="darkgrid")   to set a grid
sns.relplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=df) #hue of spliting based on category
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",style="time",data=df)
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",size="size",data=df) # size- size of that data point
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",hue="time",data=df)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",hue="day",style="time",data=df)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",hue="day",style="time",size="size",data=df)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",hue="day",size="size",palette="YlGnBu",data=df)
plt.show()
```
